```python
# Selecting only Features to use for Modeling
data = data[['we','ipron', 'pronoun', 'health', 'bio', 'affect', 'posemo', 'present', 'verb', 'anger', 'Label']] #// 'percept','hear','relativ','time','we','auxverb','future','swear','anger','number'
#feature_data = data[['pronoun','ipron', 'affect', 'posemo', 'bio','we','Label']]
#feature_data = data[['Unique', 'Dic', 'Sixltr', 'funct', 'pronoun', 'ppron','we','ipron','verb','negemo','social','health','bio', 'affect','posemo', 'family', 'present', 'work', 'verb', 'anger', 'Label']]
data
```

```python
#split-out validation dataset
array = data.values
X = array[:,0:9] #// Selecting Column //
Y = array[:,10] 
```

```python
#train/test set
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, test_size=0.2, random_state=12345)
```

```python
#KNN Classifiers
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(X_train, Y_train)
predictions = knn.predict(X_test)
print("Accuracy Score:", accuracy_score(Y_test, predictions))
#print(confusion_matrix(Y_test, predictions))
#print(classification_report(Y_test, predictions))
predictions_proba = knn.predict_proba(X_test)
print("ROC_AUC_Score:", roc_auc_score(Y_test,predictions_proba,multi_class="ovr"))
print("MCC:", matthews_corrcoef(Y_test, predictions))
print("Precision Score:",precision_score(Y_test, predictions, average='weighted'))
print("f1 Score:", f1_score(Y_test, predictions, average="micro"))
```
